# Model Karar Analizi ve İş Etkisi

## Problem Bağlamı

Bu çalışma kapsamında audiobook platformu kullanıcılarının tekrar satın alma davranışı (**Repeat Customer**) tahmin edilmiştir.

Veri setinde sınıf dağılımının dengesiz olması nedeniyle (Repeat Customer oranı yaklaşık %16), yalnızca **accuracy** metriğine odaklanan modeller iş problemi açısından yanıltıcı sonuçlar üretebilir.

Bu nedenle model performansı aşağıdaki kriterler açısından değerlendirilmiştir:

* Recall (Repeat Customer yakalama oranı)
* Precision (Yanlış hedefleme maliyeti)
* Accuracy (Genel performans)
* ROC–AUC skoru

---

## İş Hedefi (Business Objective)

Bu projede temel amaç:

> Tekrar satın alma potansiyeli olan müşterileri mümkün olduğunca kaçırmamaktır.

Ancak aşırı agresif modeller gereksiz pazarlama maliyetlerine neden olabilir.

Bu nedenle model seçiminde:

**Recall artışı ile Precision kaybı arasında dengeli bir karar stratejisi hedeflenmiştir.**

---

## Karar Stratejisi Karşılaştırması

| Model                  | Accuracy | Recall (Class 1) | Precision (Class 1) | Model Davranışı |
| ---------------------- | -------: | ---------------: | ------------------: | --------------- |
| Base (t = 0.50)        |     0.91 |             0.45 |                0.94 | Conservative    |
| Class Weight           |     0.86 |             0.74 |                0.55 | Aggressive      |
| Threshold (t = 0.40)   |     0.84 |             0.79 |                0.49 | Very Aggressive |
| ✅ Threshold (t = 0.60) |     0.88 |             0.66 |                0.60 | Balanced        |

---

## Model Seçim Gerekçesi

Daha düşük threshold değerleri repeat customer yakalama oranını artırsa da yanlış pozitif tahminleri önemli ölçüde artırmaktadır.

Bu çalışmada:

* Precision performansında ciddi kayıp oluşturmayan,
* Recall değerini anlamlı şekilde artıran,
* Pazarlama maliyetlerini kontrol altında tutan

**t = 0.60 threshold değeri nihai model olarak seçilmiştir.**

Bu yaklaşım modelin daha dengeli bir karar mekanizması üretmesini sağlamaktadır.

---

## ROC – AUC Değerlendirmesi

Modelin sınıfları ayırt etme performansı ROC eğrisi ve AUC metriği ile analiz edilmiştir.

**ROC AUC Score ≈ 0.91**

Bu sonuç modelin repeat customer ve non-repeat customer sınıflarını başarılı şekilde ayırabildiğini göstermektedir.

---

## Feature Importance (Model Yorumlanabilirliği)

Permutation Importance analizi sonucunda bazı müşteri davranışı değişkenlerinin model kararında daha etkili olduğu gözlemlenmiştir.

Bu analiz modelin yalnızca tahmin üretmekle kalmayıp:

* müşteri davranış kalıplarını öğrenebildiğini,
* iş birimi açısından yorumlanabilir içgörüler sağlayabildiğini

göstermektedir.

---

## Sonuç

Bu proje kapsamında geliştirilen model:

* sınıf dengesizliği dikkate alınarak değerlendirilmiş,
* farklı karar stratejileri test edilmiş,
* business objective odaklı threshold optimizasyonu uygulanmıştır.

Sonuç olarak model:

> Repeat customer tespitini artırırken gereksiz pazarlama maliyetlerini kontrol eden dengeli bir karar sistemi sunmaktadır.
